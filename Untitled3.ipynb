# ***TRAAJANDO CON IMG***

```python
from os import path
import cv2
import numpy as np
import matplotlib.pyplot as plt

path = "/content/images.jpg"
img = cv2.imread(path)
# Display original image
plt.figure()
plt.imshow(img)
plt.title('BGR Image')

# Display RGB image
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(img)
plt.title('Original Image')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blanco_negro=plt.imshow(img_gray, cmap='gray')
plt.figure()
plt.imshow(img)
plt.title('Blanco Negro')





```
